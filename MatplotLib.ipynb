```python
import numpy as np
import matplotlib.pyplot as plt

# Given arrays
x = [1, 2, 3, 4, 5, 6, 7, 9, 10]
y = [10, 12, 14, 16, 18, 20, 22, 26, 28]

# Plot the graph
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph between x and y')
plt.grid(True)

# Predict the range of y for x = 8
x_predict = 8
index = np.searchsorted(x, x_predict)
if index == 0:
    y_lower = y[0] - 10
    y_upper = y[0]
elif index == len(x):
    y_lower = y[-1]
    y_upper = y[-1] + 10
else:
    y_lower = y[index - 1]
    y_upper = y[index]

# Make sure the range values are multiples of 10
y_lower = (y_lower // 10) * 10
y_upper = ((y_upper + 10) // 10) * 10

print("Predicted range of y when x = 8:", y_lower, "-", y_upper)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,40,2)
y = 2 ** x
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph between x and y')
plt.plot(x,y,"b--")
plt.grid(True)
plt.show()
# Threshold for slight increase
threshold = 1

# Finding the index of the first value where there is a slight increase in y
index = np.argmax(np.diff(y) > threshold)

# Value of x where there is a slight increase in y
value_of_x = x[index + 1]

print("Value of x from where there is a slight increase in y:", value_of_x)
```

```python
x = np.arange(0,5,0.1)
y = x ** 3
y1 = x ** 2
plt.plot(x,y,color ="black", linewidth= 5,label="x^3")
plt.plot(x,y1,color ="red", linewidth= 5 ,label = "x^2")
plt.ylabel("y")
plt.xlabel('x')
# plt.axis([0,10,0,200])
plt.grid(True)
plt.text(2,80,"Text",fontsize = 12)
plt.legend()
plt.title('Matplotlib demo')
plt.show()
```

### Bubble Graph

```python
import matplotlib.pyplot as plt
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
color = ["red","black","blue","green","yellow","orange"]
plt.scatter(year,salary,s = population,c = color,alpha = 0.5,edgecolor='black',marker='^')
for i in range (len(year)):
    plt.text(year[i] + 0.1,salary[i] + 0.1,population[i])
plt.show()
```
