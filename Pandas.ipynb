### Q1 Terror Attack City

```python
import pandas as pd

# Load the dataset
terror = pd.read_csv('Dataset/terrorismData.csv')

# Filter the dataframe for incidents in Jammu and Kashmir
jk_df = terror[terror['State'] == 'Jammu and Kashmir']

# Group by city and terrorist group, and count the number of attacks
grouped = jk_df.groupby(['City', 'Group']).size().reset_index(name='Attack_Count')

# Find the city with the highest number of attacks
most_dangerous_city = grouped.loc[grouped['Attack_Count'].idxmax()]

# Print the most dangerous city and the terrorist group most active in that city
print("Most dangerous city in Jammu and Kashmir:", most_dangerous_city['City'])
print("Terrorist group most active in that city:", most_dangerous_city['Group'])
print("Number of attacks in that city:", most_dangerous_city['Attack_Count'])
```

### Q - 2 country with most attack and in which year most attack happended

```python
import pandas as pd

# Read the dataset
data = pd.read_csv('Dataset/terrorismData.csv')
df = data.copy()

# Get unique countries
unique_countries = df['Country'].unique()

# Dictionary to store the count of attacks for each country
country_attacks = {}
# Dictionary to store the year with the most attacks for each country
most_attacks_year = {}

# Count the number of attacks and find the year with the most attacks for each country
for country in unique_countries:
    country_data = df[df['Country'] == country]
    country_attacks[country] = country_data.shape[0]
    most_attacks_year[country] = country_data['Year'].value_counts().idxmax()

# Find the country with the most attacks
most_attacks_country = max(country_attacks, key=country_attacks.get)

# Print the country with the most attacks
print("Country with most attacks:", most_attacks_country)
print("Number of attacks:", country_attacks[most_attacks_country])
print("Year with most attacks in that country:", most_attacks_year[most_attacks_country])
```

```python
import pandas as pd

# Read the dataset
data = pd.read_csv('Dataset/terrorismData.csv')

# Filter out rows with invalid days (day == 0) and months (month == 0)
data = data[(data['Day'] != 0) & (data['Month'] != 0)]

# Combine day, month, and year columns into a single datetime column
data['Date'] = pd.to_datetime(data[['Day', 'Month', 'Year']])

# Filter for attacks in India after 26th May 2014
india_attacks_after_2014 = data[(data['Country'] == 'India') & (data['Date'] >= '2014-05-26')]

# Group by terrorist group and count the number of attacks for each group
grouped = india_attacks_after_2014.groupby('Group').size().reset_index(name='Attack_Count')

# Find the most active terrorist group
most_active_group = grouped.loc[grouped['Attack_Count'].idxmax()]

# Print the total number of attacks in India after 26th May 2014 and the most active terrorist group
print("Total number of attacks in India from 26th May 2014 till now:", india_attacks_after_2014.shape[0])
print("Most active terrorist group in India during this period:", most_active_group['Group'])
print("Number of attacks by this group:", most_active_group['Attack_Count'])
```

```python
import pandas as pd

# Read the dataset
data = pd.read_csv('Dataset/terrorismData.csv')

# Define the Red Corridor states
red_corridor_states = ['Jharkhand', 'Odisha', 'Andhra Pradesh', 'Chhattisgarh', 'Madhya Pradesh', 'Maharashtra', 'Uttar Pradesh', 'Telangana']

# Initialize a dictionary to store casualties for each Red Corridor state
red_corridor_casualties = {}

# Calculate casualties (sum of killed and wounded) for each Red Corridor state
for state in red_corridor_states:
    # Filter for attacks in the current state
    state_attacks = data[data['State'] == state]
    # Calculate casualties for the current state
    state_casualties = state_attacks['Killed'].sum() + state_attacks['Wounded'].sum()
    # Store casualties for the current state in the dictionary
    red_corridor_casualties[state] = state_casualties

# Print casualties for each Red Corridor state
for state, casualties in red_corridor_casualties.items():
    print("Casualties in", state, ":", casualties)
```
